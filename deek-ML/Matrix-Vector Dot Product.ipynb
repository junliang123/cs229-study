```python
import numpy as np
def matrix_dot_vector(a: list[list[int|float]], b: list[int|float]) -> list[int|float]:
	A = np.array(a)
	B = np.array(b).reshape(-1, 1)
	if A.shape[1] != B.shape[0]:
		return -1
	result = np.dot(A, B)
	return result.flatten().tolist()
print(matrix_dot_vector([[1, 2, 3], [2, 4, 5], [6, 8, 9]], [1, 2, 3]))
```
